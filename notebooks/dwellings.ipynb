```python
import pandas as pd
```

```python
dwellings = pd.read_csv(
    "https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?file=data/cens_11dwob_r3.tsv.gz",
    delimiter="\t", index_col=0
)
dwellings = dwellings.where(dwellings != ': ', other=0).astype(int)
```

```python
dwellings.where(dwellings == '70 u').dropna(axis=1, how='all')
```

```python
dwellings.loc['DW,RES1,NR,2011'].div(dwellings.loc['DW,RES2,NR,2011'].add(dwellings.loc['DW,RES_GE3,NR,2011']))
```
